A seguir estão listados três limites. Sua tarefa será encontar o valor destes limites de duas formas: algébricamente e gráficamente.

a) Para a solução algébrica, você deve explicitar todas as passagens necessárias em latex, em uma célula de texto.

b) Para solucionar graficamente deverá usar uma tabela de aproximação pela esquerda e pela direita. Lembre-se de plotar o gráfico, os pontos usados para a aproximação e de imprimir a tabela no terminal. 


1. $$\lim_{{x \to 0}} \frac{1 - \cos(2x)}{x^2}$$
---
2. $$\lim_{{x \to \infty}} \frac{3x^2 - 2x + 1}{2x^2 + x - 1}$$
---
3. $$\lim_{{x \to 0}} x \cdot \ln(1 + \sin(\frac{1}{x}))$$


FUNÇÃO 1

$\lim_{{x \to 0}} \frac{1 - \cos(2x)}{x^2}$

Aplicar a regra de L'Hôpital

$\lim_{{x \to 0}} \frac{2sin(2x)}{x^2}$

Aplicar a regra de L'Hôpital de novo

$\lim_{{x \to 0}} \frac{4cos(2x)}{2}$

Inserir o valor de x = 0

$\frac{4cos (2.0)}{2}$

Resultado é 2

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-20, 20, 1000)
y = (1 - np.cos(2*x))/x**2

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) = (1 - cos(2x))/x^2')

x_left = [-0.1, -0.01, -0.001]
y_left = [1.9983, 1.99998, 1.9999998]
plt.scatter(x_left, y_left, color='r')

x_right = [0.001, 0.01, 0.1]
y_right = [1.9999998, 1.99998, 1.9983]
plt.scatter(x_right, y_right, color='b')

plt.show()



print('Tabela de aproximação pela esquerda:')
print(' x      | f(x)')
print('--------|--------')
for x, y in zip(x_left, y_left):
    print(f'{x:<7} | {y:.7f}')



print('\nTabela de aproximação pela direita:')
print(' x      | f(x)')
print('--------|--------')
for x, y in zip(x_right, y_right):
    print(f'{x:<7} | {y:.7f}')

```

FUNÇÃO 2

Dividir pelo denominador de maior potencia:
$\lim _{x\to \infty \:}\left(\frac{3-\frac{2}{x}+\frac{1}{x^2}}{2+\frac{1}{x}-\frac{1}{x^2}}\right)$

passo 2:

$\frac{\lim _{x\to \infty \:}\left(3-\frac{2}{x}+\frac{1}{x^2}\right)}{\lim _{x\to \infty \:}\left(2+\frac{1}{x}-\frac{1}{x^2}\right)}$

passo 3:
$\lim _{x\to \infty \:}\left(3-\frac{2}{x}+\frac{1}{x^2}\right)=3$

passo 4:
$\lim _{x\to \infty \:}\left(2+\frac{1}{x}-\frac{1}{x^2}\right)=2$

resposta = $\frac{3}{2}$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)
y = (3*x**2 - 2*x + 1)/(2*x**2 + x - 1)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) = (3x^2 - 2x + 1)/(2x^2 + x - 1)')
plt.show()


x = np.array([10, 100, 1000, 10000])
y = (3*x**2 - 2*x + 1)/(2*x**2 + x - 1)


print('Tabela de aproximação:')
print(' x       | f(x)')
print('---------|--------')
for xi, yi in zip(x, y):
    print(f'{xi:<8} | {yi:.4f}')
```

 FUNÇÃO 3
 
 $\lim_{{x \to 0}} x \cdot \ln(1 + \sin(\frac{1}{x}))$

Reescrever a função como uma fração:

$\lim_{{x \to 0}} x \cdot \ln(1 + \sin(\frac{1}{x})) = \lim_{{x \to 0}} \frac{\ln(1 + \sin(\frac{1}{x}))}{\frac{1}{x}}$

Regra de L’Hôpital

$\lim_{{x \to 0}} \frac{\ln(1 + \sin(\frac{1}{x}))}{\frac{1}{x}} = \lim_{{x \to 0}} \frac{\frac{d}{dx}[\ln(1 + \sin(\frac{1}{x}))]}{\frac{d}{dx}[\frac{1}{x}]}$

Calcular as derivadas

$\lim_{{x \to 0}} \frac{\frac{d}{dx}[\ln(1 + \sin(\frac{1}{x}))]}{\frac{d}{dx}[\frac{1}{x}]} = \lim_{{x \to 0}} \frac{\cos(\frac{1}{x})\cdot(-\frac{1}{x^2})}{-1\cdot x^{-2}}$

Simplificando

$\lim_{{x \to 0}} \cos(\frac{1}{x})$

Resposta

O limite não existe, pois o "cos(1/x)" oscila entre -1 e 1 quando x se aproxima de 0

```python
import matplotlib.pyplot as plt
import numpy as np

x_left = np.linspace(-0.1, -0.0001, 1000)
y_left = x_left * np.log(1 + np.sin(1/x_left))

x_right = np.linspace(0.0001, 0.1, 1000)
y_right = x_right * np.log(1 + np.sin(1/x_right))

plt.plot(x_left, y_left)
plt.plot(x_right, y_right)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) = x * ln(1 + sin(1/x))')
plt.show()


import numpy as np


x_left = np.array([-0.1, -0.01, -0.001])
y_left = x_left * np.log(1 + np.sin(1/x_left))


print('Tabela de aproximação pela esquerda:')
print(' x      | f(x)')
print('--------|--------')
for x, y in zip(x_left, y_left):
    print(f'{x:<7} | {y:.7f}')


x_right = np.array([0.001, 0.01, 0.1])
y_right = x_right * np.log(1 + np.sin(1/x_right))


print('\nTabela de aproximação pela direita:')
print(' x      | f(x)')
print('--------|--------')
for x, y in zip(x_right, y_right):
    print(f'{x:<7} | {y:.7f}')
```

# Questão 2
Considerando a função definida por: 
A função descontínua $f(x)$ é definida como:

$$
f(x) = 
\begin{cases} 
\sin(x) & \text{se } -10 \leq x < -3 \\
\cos(x) & \text{se } -3 \leq x < 0 \\
x^2 & \text{se } 0 \leq x < 3 \\
e^{x-10} & \text{se } 3 \leq x \leq 10
\end{cases}
$$

Trace o gráfico desta função, marque os pontos de descontinuidade, caso existam estes pontos, com um círculo vermelho, anote estes pontos. Não se esqueça de colocar uma legenda que permita o entendimento do seu gráfico. 

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-10, -3, 100)
y1 = np.sin(x1)

x2 = np.linspace(-3, 0, 100)
y2 = np.cos(x2)

x3 = np.linspace(0, 3, 100)
y3 = x3**2

x4 = np.linspace(3, 10, 100)
y4 = np.exp(x4-10)

plt.plot(x1, y1, 'b', label='sin(x) se -10 <= x < -3')
plt.plot(x2, y2, 'g', label='cos(x) se -3 <= x < 0')
plt.plot(x3, y3, 'r', label='x^2 se 0 <= x < 3')
plt.plot(x4, y4, 'c', label='e^(x-10) se 3 <= x <= 10')

plt.plot([-3,-3],[-1.5,1.5],'k--')
plt.plot([0,0],[-1.5,1.5],'k--')
plt.plot([3,3],[-1.5,np.exp(-7)],'k--')

plt.scatter(-3,np.cos(-3),s=50,c='red',zorder=10)
plt.scatter(0,(0)**2,s=50,c='red',zorder=10)
plt.scatter(3,np.exp(3-10),s=50,c='red',zorder=10)

plt.legend(loc='best')
plt.ylim([-1.5,np.exp(-7)+0.5])
plt.xlim([-10.5,10.5])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x)')
plt.show()
```
